```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#提取数据集
url = 'G:\DownLoad\data\production_missing_num.csv'
df_production = pd.read_csv(url)
```

```python
#将thickness值转换为float型，并将字符型None转换为null
df_production['thickness'] = df_production['thickness'].replace('None', np.nan)
df_production['thickness'] = df_production['thickness'].astype(np.float64)

#将falut_flg的类型转换为布尔类型
df_production['fault_flg'] = df_production['fault_flg'].replace('False', 0)
df_production['fault_flg'] = df_production['fault_flg'].replace('True', 1)
df_production['fault_flg'] = df_production['fault_flg'].astype(np.int64)
```

```python
#查看thickness中是否有缺失值
print(df_production['thickness'].isnull().any(),'\n')
#显示有缺失值的地方
print(df_production[df_production['thickness'].isnull().values==True])
```

```python
#分别查看产品A,B,C,D,E的平均值
df_production.groupby('type').mean()
```

```python
#显示df_production的数据统计量
df_production.describe()
#做length和thickness的散点图
plt.scatter(df_production['length'], df_production['thickness'], alpha=0.8)
```

```python
#通过数据分组遍历各组数据
type_group = df_production.groupby('type')
i = 0

for name, group in type_group:
    i = i+1
    print('Group', i, name)
    print(group)
```

```python
#通过产品种类分为A,B,C,D,E组，并将他们单独提取出来

group = df_production.groupby('type')
df_A = type_group.get_group(('A'))[['type', 'length', 'thickness', 'fault_flg']]
df_B = type_group.get_group(('B'))[['type', 'length', 'thickness', 'fault_flg']]
df_C = type_group.get_group(('C'))[['type', 'length', 'thickness', 'fault_flg']]
df_D = type_group.get_group(('D'))[['type', 'length', 'thickness', 'fault_flg']]
df_E = type_group.get_group(('E'))[['type', 'length', 'thickness', 'fault_flg']]
```

```python
#计算各产品的均值，用于填充缺失值
mean_A = df_A['thickness'].mean()
mean_B = df_B['thickness'].mean()
mean_C = df_C['thickness'].mean()
mean_D = df_D['thickness'].mean()
mean_E = df_E['thickness'].mean()

df_A['thickness'] = df_A['thickness'].replace(np.nan,mean_A)
df_B['thickness'] = df_B['thickness'].replace(np.nan,mean_B)
df_C['thickness'] = df_C['thickness'].replace(np.nan,mean_C)
df_D['thickness'] = df_D['thickness'].replace(np.nan,mean_D)
df_E['thickness'] = df_E['thickness'].replace(np.nan,mean_E)
```

```python
print(df_A.isnull().any(),'\n')
print(df_B.isnull().any(),'\n')
print(df_C.isnull().any(),'\n')
print(df_D.isnull().any(),'\n')
print(df_E.isnull().any(),'\n')
#全为false，说明缺失值被填充
```

```python
#画箱线图
plt.figure()
df_A.boxplot()

plt.figure()
df_B.boxplot()

plt.figure()
df_C.boxplot()

plt.figure()
df_D.boxplot()

plt.figure()
df_E.boxplot()
```

```python
#标准差检测异常值
from scipy.stats import zscore

#A产品的length异常值:
print('A类产品的length异常值：')
df_A['length'][np.abs(zscore(df_A['length']))>3]

#B产品的length异常值:
print('B类产品的length异常值：')
df_B['length'][np.abs(zscore(df_B['length']))>3]

#C产品的length异常值:
print('C类产品的length异常值：')
df_C['length'][np.abs(zscore(df_C['length']))>3]

#D产品的length异常值:
print('D类产品的length异常值：')
df_D['length'][np.abs(zscore(df_D['length']))>3]

#E产品的length异常值:
print('E类产品的length异常值：')
df_E['length'][np.abs(zscore(df_E['length']))>3]

#A产品的thickness异常值：
print('A类产品的thickness异常值：')
df_A['thickness'][np.abs(zscore(df_A['thickness']))>3]

#B产品的thickness异常值：
print('B类产品的thickness异常值：')
df_B['thickness'][np.abs(zscore(df_B['thickness']))>3]

#C产品的thickness异常值：
print('C类产品的thickness异常值：')
df_C['thickness'][np.abs(zscore(df_C['thickness']))>3]

#D产品的thickness异常值：
print('D类产品的thickness异常值：')
df_D['thickness'][np.abs(zscore(df_D['thickness']))>3]

#E产品的thickness异常值：
print('E类产品的thickness异常值：')
df_E['thickness'][np.abs(zscore(df_E['thickness']))>3]
```

```python
#将A,B,C,D,E五个数据集整合在一起
df_product = pd.concat([df_A, df_B, df_C, df_D, df_E], axis = 0)
df_product.describe()
```

```python
#没有发现异常点
#选择用Z-score将数据标准化

def ZscoreNormalization(x):
    x = (x - np.mean(x))/np.std(x)
    return x

Z_views = pd.DataFrame()
Z_views['zscore_length'] = ZscoreNormalization(df_product['length'])
Z_views['zscore_thickness'] = ZscoreNormalization(df_product['thickness'])
Z_views['fault_flg'] = df_product['fault_flg']
print(Z_views)
plt.scatter(Z_views['zscore_length'], Z_views['zscore_thickness'], alpha=0.8)
```

```python
#PCA降维
from sklearn.decomposition import PCA
pca = PCA(n_components = 1)
pca.fit(Z_views)
new_product = pca.fit_transform(Z_views)
print(pca.explained_variance_ratio_)
print(new_product)
#重构
reconMat = pca.inverse_transform(new_product)
```

```python
plt.scatter(Z_views['zscore_length'], Z_views['zscore_thickness'], alpha=0.8)
plt.scatter(reconMat[:,0], reconMat[:,1], alpha=0.8,color = 'red')
```
